```python
import pymcel as pc
import numpy as np
import matplotlib.pyplot as plt
```

Trayectoria del vector relativo en el sistema perifocal:

```python
p = 1
e = 0.7
thetars = np.linspace(0, 2 * np.pi, 100)

rs = p / (1 + e * np.cos(thetars))

xfs = rs * np.cos(thetars)
yfs = rs * np.sin(thetars)

plt.plot(xfs, yfs)
plt.axis('equal')
plt.grid()
plt.tight_layout()
```

Transformación de Jacobi:

```python
m1 = 1
m2 = 1
M = m1 + m2

rvecs = np.array([xfs, yfs]).T

rlvecs = m2 / M * rvecs
r2vecs = -m1 / M * rvecs
```

Graficamos las trayectorias:

```python
plt.plot(rlvecs[:,0], rlvecs[:,1], label='Cuerpo 1')
plt.plot(r2vecs[:,0], r2vecs[:,1], label='Cuerpo 2')
plt.legend()
plt.axis('equal')
plt.grid()
plt.tight_layout()
```
