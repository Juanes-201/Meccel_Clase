# Punto 2. Magnetósferas de los Gigantes.

## Introducción 
El tamaño de la magnetósfera de un planeta está determinado por el equilibrio entre la presión ejercida por el viento solar y la presión magnética generada por el campo dipolar planetario. Este equilibrio define la distancia de standoff, que corresponde a la posición de la magnetopausa en la dirección del Sol. En este análisis, se emplea una relación empírica que incorpora tanto las propiedades intrínsecas del planeta (como su momento magnético), como las condiciones locales del viento solar.

```python
def magneto(M,n, T):
    """
    M: Momento magnetico del planeta (in A m^2)
    n: Densidad numerica del viento solar (en particulas por m^3)
    T: temperatura del viento solar (en K)

    Retorna la distancia de standoff de la magnetopausa en metros.
    """
    # Constantes

    # Permeabilidad del vacío  (en T m/A)
    mu_0 = 4 * np.pi * 1e-7 

    # Constante de Boltzmann (en J/K)
    k_B = 1.38e-23  
    
    # Presion dinamica del viento solar
    P_vs =2 * n * k_B * T  # en Pascales
    
    f_0 = 1.66  # Factor de forma

    #Masa del proton
    m = 1.67e-27  # en kg

    # Velocidad del viento solar promedio

    v = 400e3  # en m/s (400 km/s)
    # Distancia de standoff de la magnetopausa
    R_mp = ( mu_0 * f_0**2 * M**2 / (8 * np.pi**2 * (m*n*v**2 + P_vs) ))**(1/6)  # en metros
    
    return R_mp
```

```python
Planetas=[{"nombre": "Jupiter", "M": 1.56e27, "n": 1.85e5, "T": 3.16e4}, {"nombre": "Saturno", "M": 4.6e25, "n": 5.54e4, "T": 2.07e4}, {"nombre": "Urano", "M": 3.9e24, "n": 1.36e4, "T": 1.26e4}, {"nombre": "Neptuno", "M": 2.2e24, "n": 5.52e3, "T": 9.2e3}]
```

```python
Jup=magneto(Planetas[0]["M"], Planetas[0]["n"], Planetas[0]["T"])
Sat=magneto(Planetas[1]["M"], Planetas[1]["n"], Planetas[1]["T"])
Ur=magneto(Planetas[2]["M"], Planetas[2]["n"], Planetas[2]["T"])
Nep=magneto(Planetas[3]["M"], Planetas[3]["n"], Planetas[3]["T"])


print(f"La distancia standoff de la magnetósfera de Júpiter es: {Jup:.3e} metros")
print(f"La distancia standoff de la magnetósfera de Saturno es: {Sat:.3e} metros")
print(f"La distancia standoff de la magnetósfera de Urano es: {Ur:.3e} metros")
print(f"La distancia standoff de la magnetósfera de Neptuno es: {Nep:.3e} metros")
```

## Conclusión

Los resultados obtenidos reproducen correctamente el orden de magnitud de las distancias de standoff reportadas en la literatura, mostrando una buena concordancia general con los valores observacionales. Las diferencias observadas pueden atribuirse a las simplificaciones del modelo, el cual asume un campo dipolar ideal y condiciones estacionarias del viento solar, sin considerar efectos adicionales como la presencia de plasma interno, la formación de magnetodiscos o la variabilidad temporal del entorno solar. En conjunto, el modelo empleado resulta adecuado para estimaciones de primer orden dentro del contexto de ciencias planetarias.
