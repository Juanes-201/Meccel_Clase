```python
import pymcel as pc
import numpy as np
deg = np.pi / 180
import matplotlib.pyplot as plt
```

```python
e = 0.1911663355386932	
a =	0.9223803173917017 * pc.constantes.au
q =	0.7460522521429133 * pc.constantes.au
i =	3.340958441017069*deg
node =	203.8996515621043*deg	
peri =	126.6728325163065*deg
tp =	2461042.918242006079 * 86400 # Fecha Juliana
mu = pc.constantes.mu_sun # Este es el mu de todo el sistema solar. Lo aproximamos a la masa del Sol.
```

Variables Derivadas

```python
h = np.sqrt(mu * a * (1-e**2))
b = a * np.sqrt(1 - e**2)
h, b
```

¿Cuál es el tiempo?

```python
from astropy.time import Time
t = Time("2029-04-13 18:52:00", scale="tdb").jd * 86400 # Convertimos a segundos
t
```

Escribir la ecuación en la forma f(E)=0

```python
def ecuacion_kepler(E):
    funcion = E - e * np.sin(E) - h/(a * b) * (t - tp) 
    return funcion
```

Resolver la ecuación de Kepler

```python
from scipy.optimize import newton
```

```python
newton(ecuacion_kepler, 0) #Faltan cositas
```
